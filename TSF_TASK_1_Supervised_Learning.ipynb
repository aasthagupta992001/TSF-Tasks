# **Supervised Learning**

Supervised learning is the types of machine learning in which machines are trained using well "labelled" training data, and on basis of that data, machines predict the output. The labelled data means some input data is already tagged with the correct output. Supervised learning is a process of providing input data as well as correct output data to the machine learning model.

## **Importing required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Handling the Data**

```python
df = pd.read_csv('student_scores.csv')
df.head()
```

## **Visualisations**

```python
df.plot(x='Hours', y='Scores', style='*')
plt.title('Hours Vs Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
sns.regplot(x="Hours", y="Scores", data=df)
```

## **Training the Model**

```python
X= df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
plt.scatter(X,y)
```

```python
print("Comparison between Actual Vs Prediction")
data = pd.DataFrame({'Actual':y_test, 'Predicted':pred})
data
```

## **Predicting the Value as given in the Task**

```python
# predicting for 9.25 hours
model.predict([[9.25]])
```

```python
from sklearn.metrics import mean_absolute_error
print('Mean Absolute Error:',mean_absolute_error(y_test,pred))
```
